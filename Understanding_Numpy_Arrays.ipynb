## The fourth dimension
Printing arrays is a good way to check code output for small arrays like sudoku_game_and_solution, but it becomes unwieldy when dealing with bigger arrays and those with higher dimensions. Another important check is to look at the array's .shape.

Now, we'll create a 4D array that contains two sudoku games and their solutions. 

```python
new_sudoku_game = np.array([[0, 0, 4, 3, 0, 0, 0, 0, 0],
       [8, 9, 0, 2, 0, 0, 6, 7, 0],
       [7, 0, 0, 9, 0, 0, 0, 5, 0],
       [5, 0, 0, 0, 0, 8, 1, 4, 0],
       [0, 7, 0, 0, 3, 2, 0, 6, 0],
       [6, 0, 0, 0, 0, 1, 3, 0, 8],
       [0, 0, 1, 7, 5, 0, 9, 0, 0],
       [0, 0, 5, 0, 4, 0, 0, 1, 2],
       [9, 8, 0, 0, 0, 6, 0, 0, 5]])
```

```python
new_sudoku_solution = np.array([[2, 5, 4, 3, 6, 7, 8, 9, 1],
       [8, 9, 3, 2, 1, 5, 6, 7, 4],
       [7, 1, 6, 9, 8, 4, 2, 5, 3],
       [5, 3, 2, 6, 9, 8, 1, 4, 7],
       [1, 7, 8, 4, 3, 2, 5, 6, 9],
       [6, 4, 9, 5, 7, 1, 3, 2, 8],
       [4, 2, 1, 7, 5, 3, 9, 8, 6],
       [3, 6, 5, 8, 4, 9, 7, 1, 2],
       [9, 8, 7, 1, 2, 6, 4, 3, 5]])
```

```python
# Create a second 3D array of another game and its solution 
new_game_and_solution = np.array([new_sudoku_game,new_sudoku_solution])

# Create a 4D array of both game and solution 3D arrays
games_and_solutions = np.array([game_and_solution, new_game_and_solution])

# Print the shape of your 4D array
print(games_and_solutions.shape)
```

Fabulous 4D array! The .shape of (2, 2, 9, 9) means that there are two sets of game/solution pairs. Within each game/solution pair, all arrays have nine rows and nine columns. That's exactly right! If there were ten game/solution pairs, the .shape would be (10, 2, 9, 9).


## Flattening and reshaping
We'll change sudoku_game from a 2D array to a 1D array and back again. Can we trust NumPy to keep the array elements in the same order after being flattened and reshaped? Time to find out.

```python
# Flatten sudoku_game
flattened_game = sudoku_game.flatten()

# Print the shape of flattened_game
print(flattened_game.shape)
```

```python
# Reshape flattened_game back to a nine by nine array
reshaped_game = flattened_game.reshape((9,9))

# Print sudoku_game and reshaped_game
print(sudoku_game)
print(reshaped_game)
```

Notice that .flatten() preserves the order of array elements—when the array is reshaped back to its original shape, the numbers are in the exact same places as they were before flattening.

## The dtype argument
One way to control the data type of a NumPy array is to declare it when the array is created using the dtype keyword argument.

```python
# Create an array of zeros with three rows and two columns
zero_array = np.zeros((3,2))

# Print the data type of zero_array
print(zero_array.dtype)
```

```python
# Create a new array of int32 zeros with three rows and two columns
zero_int_array = np.zeros((3,2), dtype = np.int32)

# Print the data type of zero_int_array
print(zero_int_array.dtype)
```

Nice! Since our original zero_array did not have any information following the decimal places of the floats, it makes sense to change the array's data type to an integer with a smaller bitsize.

## A smaller sudoku game
NumPy data types, which emphasize speed, are more specific than Python data types, which emphasize flexibility. When working with large amounts of data in NumPy, it's good practice to check the data type and consider whether a smaller data type is large enough for our data, since smaller data types use less memory.

It's time to make our sudoku game more memory-efficient using data types!

```python
# Print the data type of sudoku_game
print(sudoku_game.dtype)
```

```python
# Change the data type of sudoku_game to int8
small_sudoku_game = sudoku_game.astype('int8')

# Print the data type of small_sudoku_game
print(small_sudoku_game.dtype)
```

int8 is one of the smallest data types in NumPy since 8 bits is only a single byte.
